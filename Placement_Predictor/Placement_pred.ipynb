```python
import numpy as np 
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20,8

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("Placement_Data_Full_Class.csv")
```

```python
gen = { 'M':0, 'F':1}
data["gender"] = data["gender"].map(gen)
```

```python
board = {'Others':0,'Central':1}
data["ssc_b"]=data["ssc_b"].map(board)
data["hsc_b"]=data["hsc_b"].map(board)
```

```python
stream = {'Arts':0, 'Commerce':1, 'Science':2}
data['hsc_s']=data['hsc_s'].map(stream)
```

```python
tech_stream = {'Others':0, 'Comm&Mgmt':1, 'Sci&Tech':2}
data['degree_t']=data['degree_t'].map(tech_stream)
```

```python
exp = { 'No':0, 'Yes':1}
data["workex"] = data["workex"].map(exp)
```

```python
spe = { 'Mkt&HR':0, 'Mkt&Fin':1}
data["specialisation"] = data["specialisation"].map(spe)
```

```python
place = { 'Not Placed':0, 'Placed':1}
data["status"] = data["status"].map(place)
```

```python
data.drop("sl_no",inplace=True,axis=1)
```

```python
sns.factorplot(x="workex",y="status",data=data)
```

```python
sns.lineplot(x="mba_p", y="status", data=data)
```

```python
c = ["salary","status"]
x = data.drop(c,axis=1)
y = data["status"]
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
lrp = lr.predict(xtest)
lra = accuracy_score(lrp,ytest)
print("pecent accuracy = ",round(lra*100,2)) 
```

```python
dt =DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
dtp = dt.predict(xtest)
dta = accuracy_score(dtp,ytest)
print("pecent accuracy = ",round(dta*100,2))
```

```python
kn =KNeighborsClassifier(n_neighbors=10)
kn.fit(xtrain,ytrain)
knp = kn.predict(xtest)
kna = accuracy_score(knp,ytest)
print("pecent accuracy = ",round(kna*100,2))
```

```python
rf =RandomForestClassifier(n_estimators=70)
rf.fit(xtrain,ytrain)
rfp = rf.predict(xtest)
rfa = accuracy_score(rfp,ytest)
print("pecent accuracy = ",round(rfa*100,2))
```
